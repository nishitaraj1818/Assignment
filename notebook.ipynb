```python
import pandas as pd

df = pd.read_csv("data/superstore.csv", encoding="latin1")

df.head()
```

## Insight
- Verifies the CSV import and confirms the dataset is available for analysis.


```python
import sqlite3
```

## Insight
-  Prepares a local SQL database for querying the dataset.

```python
# Create SQLite DB and load table
conn = sqlite3.connect("superstore.db")
df.to_sql("orders", conn, if_exists="replace", index=False)

print("Loaded", len(df), "rows")
```

## Insight
- Persists data to SQL so later analysis can use SQL queries.

```python
# Schema
cursor = conn.cursor()
cursor.execute("PRAGMA table_info(orders)")
for col in cursor.fetchall():
    print(col)

# Sample data
pd.read_sql("SELECT * FROM orders LIMIT 5", conn)
```

## Insight
- Confirms table structure and validates data load into SQLite.

```python
# Filter by region
pd.read_sql("SELECT * FROM orders WHERE Region = 'West' LIMIT 10", conn)

# Filter by category
pd.read_sql("SELECT * FROM orders WHERE Category = 'Furniture' LIMIT 10", conn)

# Filter by date range
pd.read_sql("""
    SELECT * FROM orders 
    WHERE "Order Date" BETWEEN '2016-01-01' AND '2016-12-31' 
    LIMIT 10
""", conn)

# Filter by sales threshold
pd.read_sql("SELECT * FROM orders WHERE Sales > 500 ORDER BY Sales DESC LIMIT 10", conn)
```

## Insight
- Demonstrates basic SQL slicing for targeted data exploration.

```python
# Total sales by region
pd.read_sql("""
    SELECT Region, SUM(Sales) AS total_sales, SUM(Quantity) AS total_qty
    FROM orders
    GROUP BY Region
    ORDER BY total_sales DESC
""", conn)

# Average sales & profit by category
pd.read_sql("""
    SELECT Category, AVG(Sales) AS avg_sales, AVG(Profit) AS avg_profit
    FROM orders
    GROUP BY Category
""", conn)

# Sales by sub-category
pd.read_sql("""
    SELECT "Sub-Category", SUM(Sales) AS total_sales
    FROM orders
    GROUP BY "Sub-Category"
    ORDER BY total_sales DESC
""", conn)
```

## Insight
- Provides regional and product-level performance summaries.

```python
# Top 10 best-selling products
pd.read_sql("""
    SELECT "Product Name", SUM(Sales) AS total_sales
    FROM orders
    GROUP BY "Product Name"
    ORDER BY total_sales DESC
    LIMIT 10
""", conn)

# Top 5 categories by profit
pd.read_sql("""
    SELECT Category, SUM(Profit) AS total_profit
    FROM orders
    GROUP BY Category
    ORDER BY total_profit DESC
    LIMIT 5
""", conn)
```

## Insight
- Identifies best-performing products and high-profit categories.

```python
pd.read_sql("""
    SELECT strftime('%Y-%m', "Order Date") AS month, SUM(Sales) AS monthly_sales
    FROM orders
    GROUP BY month
    ORDER BY month
""", conn)
```

## Insight
- Reveals time-based sales trends across the dataset.

```python
pd.read_sql("""
    SELECT "Customer Name", SUM(Sales) AS total_spend
    FROM orders
    GROUP BY "Customer Name"
    ORDER BY total_spend DESC
    LIMIT 10
""", conn)
```

## Insight
- Shows the highest-value customers driving the most revenue.

```python
pd.read_sql("""
    SELECT "Order ID", "Product ID", COUNT(*) AS cnt
    FROM orders
    GROUP BY "Order ID", "Product ID"
    HAVING COUNT(*) > 1
""", conn)
```

## Insight
- Checks data quality for repeated line items in single orders.

```python
# Total row count
pd.read_sql("SELECT COUNT(*) AS total_rows FROM orders", conn)

# Null check on key columns
pd.read_sql("""
    SELECT 
        SUM(CASE WHEN Sales IS NULL THEN 1 ELSE 0 END) AS null_sales,
        SUM(CASE WHEN Region IS NULL THEN 1 ELSE 0 END) AS null_region,
        SUM(CASE WHEN "Order Date" IS NULL THEN 1 ELSE 0 END) AS null_date
    FROM orders
""", conn)
```
